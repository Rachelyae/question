# Predicting mortgage defaults


## Data Cleaning
Merged 3 files into one according to year,quater,and month.

Removed certain columns like 'loan_id','state',and add changes for treasury bill rate and current 30-year fixed mortgage rate from a year ago as'delta_F','delta_t'.

Target variable (delinquent30) - encoded to 0 or 1

```python
import pandas as pd
tm=pd.DataFrame(pd.read_csv('mortgage data.csv',low_memory=False))

def remove_f(month):
    month=str(month)[0:1]
    return int(month)
tm['month'] = tm['frst_dte'].apply(remove_f)

tm=tm.drop(['source','frst_dte'],axis=1)
tm[tm['fthb_flg']=='U'] = 'N'

th=pd.DataFrame(pd.read_csv('hpi_state.csv',low_memory=False))
th=th.drop(['state_code','state_full'],axis=1)
th=th.rename(columns={'quarter':'Quarter_orig','year':'Year_orig'})

tr=pd.read_csv('rate.csv')
tr=tr.rename(columns={'year':'Year_orig'})

t1 =pd.merge(tm,th,how='left')
t2 =pd.merge(t1,tr,how='left')

from sklearn.preprocessing import LabelEncoder
t2.fthb_flg = LabelEncoder().fit_transform(t2.fthb_flg)

t2=t2.drop(['loan_id','state',],axis=1)


col = list(t2.columns)
 
# Convert all column types to float, fill in NaN for errors, and fill mean of columns for NaN
t2[col] = t2[col].apply(pd.to_numeric, errors='coerce')
t2=t2.fillna((t2[col].mean()))
t2 = pd.DataFrame(t2, dtype='float')


t2.head(200)
```

Standardrize some features.

```python
from sklearn.preprocessing import StandardScaler
t2['orig_amt'] = StandardScaler().fit_transform(t2['orig_amt'].values.reshape(-1, 1))
t2['oltv'] = StandardScaler().fit_transform(t2['oltv'].values.reshape(-1, 1))
t2['ocltv'] = StandardScaler().fit_transform(t2['ocltv'].values.reshape(-1, 1))
t2['dti'] = StandardScaler().fit_transform(t2['dti'].values.reshape(-1, 1))
t2['cscore_b'] = StandardScaler().fit_transform(t2['cscore_b'].values.reshape(-1, 1))
t2['mi_pct'] = StandardScaler().fit_transform(t2['mi_pct'].values.reshape(-1, 1))
t2['HPI_state'] = StandardScaler().fit_transform(t2['HPI_state'].values.reshape(-1, 1))
t2['FRM30_rate'] = StandardScaler().fit_transform(t2['FRM30_rate'].values.reshape(-1, 1))
t2.head(100)
```
